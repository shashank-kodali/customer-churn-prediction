````python
import shutil, os

DRIVE_PATH     = '/content/drive/MyDrive/Churn_Hackathon/'
PROCESSED_PATH = DRIVE_PATH + 'data/processed/'
REPO_PATH      = '/content/repo/'

# ── Ensure repo folders exist ──────────────────────────────
for folder in ['notebooks', 'models', 'reports']:
    os.makedirs(REPO_PATH + folder, exist_ok=True)

# ── 1. Notebooks (auto-picks ALL .ipynb files) ─────────────
nb_dir = DRIVE_PATH + 'colab_notebooks/'
if os.path.exists(nb_dir):
    for file in os.listdir(nb_dir):
        if file.endswith('.ipynb'):
            shutil.copy(nb_dir + file, REPO_PATH + 'notebooks/' + file)
            print(f"📓 Notebook  → {file}")
else:
    print("⚠️  Notebooks folder not found")

# ── 2. Model files (auto-picks ALL .pkl files) ─────────────
if os.path.exists(PROCESSED_PATH):
    for file in os.listdir(PROCESSED_PATH):
        if file.endswith('.pkl'):
            shutil.copy(PROCESSED_PATH + file, REPO_PATH + 'models/' + file)
            print(f"🤖 Model     → {file}")
else:
    print("⚠️  Processed folder not found")

# ── 3. Report images (auto-picks ALL .png files) ───────────
data_dir = DRIVE_PATH + 'data/'
if os.path.exists(data_dir):
    for file in os.listdir(data_dir):
        if file.endswith('.png'):
            shutil.copy(data_dir + file, REPO_PATH + 'reports/' + file)
            print(f"📊 Report    → {file}")
else:
    print("⚠️  Data folder not found")

print("\n✅ All files synced to repo!")
````
